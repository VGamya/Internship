```python
match=[]
try:
    m=driver.find_elements_by_xpath("//span[@class='ng-binding']")
    for i in m:
        match.append(i.text)
except:
    match.append('-')
match
```

```python
res=[]
try:
    s=driver.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")
    for i in s:
        res.append(i.text)
except:
    res.append('-')
res
```

```python
series=[]
place=[]
for i in res:
    a=i.split('-')
    series.append(a[0])
    place.append(a[1])
    a=0
```

```python
series
```

```python
place
```

```python
date=[]
d=driver.find_elements_by_xpath("//div[@class='match-card-left match-schedule']")
for i in d:
    date.append(i.text)
date
```

```python
time=[]
t=driver.find_elements_by_xpath("//div[@class='match-card-right match-schedule ']")
for i in t:
    time.append(i.text)
time
```

```python
fixtures=pd.DataFrame({})
fixtures['Match Title']=match
fixtures['Series']=series
fixtures['Place']=place
fixtures['Date']=date
fixtures['Time Note']=time
fixtures
```
