```python
a = df['Complete, relevant description information of products'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["snapdeal.com"] = a[" Snapdeal.com"]+a["Snapdeal.com"]
a.drop([" Snapdeal.com","Snapdeal.com"],axis=1,inplace=True)

sn.swarmplot(x=a.sum(),y=a.sum(),hue=a.columns)
plt.xlabel('Complete, relevant description information of products')
plt.title("products information", fontsize=15)
plt.show()
```

238 respondents says that amazon gives the complete relevant description information of products, whereas 59 responedents feel that snapdeal and paytm gives the complete relevant description of products which is the least of all.

```python
a = df['Quickness to complete purchase'].str.get_dummies(',')
a["flipkart.com"] = a[" Flipkart.com"]+a["Flipkart.com"]
a.drop([" Flipkart.com","Flipkart.com"],axis=1,inplace=True)
a["paytm.com"] = a[" Paytm.com"]+a["Paytm.com"]
a.drop([" Paytm.com","Paytm.com"],axis=1,inplace=True)
sn.barplot(x=a.sum(),y=a.sum(),hue=a.columns,dodge=False)
plt.xlabel("Quickness to complete purchase")
plt.title("Quickness to complete purchase",fontsize=15)
plt.show()
```
